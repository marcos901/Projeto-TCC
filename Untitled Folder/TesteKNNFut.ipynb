```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importando a base de dados
df = pd.read_csv("nba.games.stats.csv")
```

```python
# Selecionando o que me interessa da base
df.columns
```

```python
df.drop(['Unnamed: 0', 'Team', 'Game', 'Date', 'Home', 'Opponent'], axis=1, inplace=True)
```

```python
#Transformando os dados categóricos de FTR para numeros
from sklearn.preprocessing import LabelEncoder
# Create a label (category) encoder object
le = LabelEncoder()
# Fit the encoder to the pandas column
le.fit(df['WINorLOSS'])
# Apply the fitted encoder to the pandas column
df['WINorLOSS'] = le.transform(df['WINorLOSS'])
```

```python
#normalizando os dados
from sklearn.preprocessing import StandardScaler
```
